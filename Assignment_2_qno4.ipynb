```python
import numpy as np
import time
def divided_difference_comp(xi, prev_divided_difference,i):
  size_div_difference=np.size(prev_divided_difference) -i       #most crucial step takes care of indexing
  divided_difference=np.zeros(size_div_difference)
  for j in range(0,size_div_difference):
    numerator=prev_divided_difference[j+1]-prev_divided_difference[j]
    denominator=xi[j+i]-xi[j]
    divided_difference[j]= numerator/denominator
  return divided_difference

def newton_polynomial(xi,divided_difference,evaluation_point):
  polynomial=divided_difference[0]                             #here divided diff contains the north-east edge point of the divided diff table
  basis_term=np.ones(np.size(xi))
  for i in range(1,np.size(xi)):
    for j in range(0,i):
      basis_term[i] *=evaluation_point - xi[j]
    polynomial += divided_difference[i]* basis_term[i]
  return polynomial

def f_1(x):
  return np.cos(x)

def f_2(x):
  return 1/(1+25*(x**2))

a= -1
b= 1

number_points_1=5
number_points_2=25
number_points_3=50

xi=np.linspace(a,b,number_points_1)
qi=np.linspace(a,b,number_points_3)
ri=np.linspace(a,b,number_points_3)

fi_1=f_1(xi)
fi_2=f_2(xi)

evaluation_point_1= 0
evaluation_point_2=0.95

analytical_solution_f_1_eva_point_1=f_1(evaluation_point_1)
analytical_solution_f_1_eva_point_2=f_1(evaluation_point_2)
analytical_solution_f_2_eva_point_1=f_2(evaluation_point_1)
analytical_solution_f_2_eva_point_2=f_2(evaluation_point_2)

# Now,forming the divided diff table:

divided_difference_f1 =np.zeros((np.size(xi),np.size(xi)))
divided_difference_f2 =np.zeros((np.size(xi),np.size(xi)))
divided_difference_f1[:, 0]=fi_1
divided_difference_f2[:, 0]=fi_2





#t=[2,4,9,10]
#r=[4,56,711,980]
#print(divided_difference_comp(t,r,1))
#p=[0,1,3,4]                           #to check the function is defined correctly or not
#q=[-5,1,25,55]
#c=[-34,-3.2,0.33333333,-0]
#d=[2,2.5,3,3.5,4]
#e=[9,26.5,15,2,0]
#print(divided_difference_comp(a,c,2))
#print(newton_polynomial(a,b,2))
#print(newton_polynomial(d,e,3.8))

#from 1st dd onwards:

for i in range(1,np.size(xi)):
  prev_dd= i-1
  divided_difference_f1[:len(xi)-i,i]=  divided_difference_comp(xi, divided_difference_f1[:,prev_dd], i)
  divided_difference_f2[:len(xi) - i, i] = divided_difference_comp(xi, divided_difference_f2[:, prev_dd], i)

#for n=5 and evaluation point 0

numerical_solution_f1=newton_polynomial(xi,divided_difference_f1[0,:],evaluation_point_1)
start_time=time.time()
numerical_solution_f2=newton_polynomial(xi,divided_difference_f2[0,:],evaluation_point_1)
end_time=time.time()

error_f1_eva_point_1=np.abs(numerical_solution_f1 - analytical_solution_f_1_eva_point_1)
error_f2_eva_point_1=np.abs(numerical_solution_f2 - analytical_solution_f_2_eva_point_1)

print(f"N_points: {number_points_1}")
print(f"Evaluation Point : {evaluation_point_1}")
print(f"f_1(x) : cos(x)")
print(f"Numerical Solution : {numerical_solution_f1} , Analytical Solution : {analytical_solution_f_1_eva_point_1} , Error : {error_f1_eva_point_1}")

print(f"f_2(x) : 1/1+25x**2")
print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f_2_eva_point_1} , Error : {error_f2_eva_point_1}")

computed_time=end_time - start_time
print(f"Elasped time for Newton Polynomial at x_eval : {evaluation_point_1} = {computed_time} seconds")

#Now,for n=5 and evaluation point 0.95

numerical_solution_f1=newton_polynomial(xi,divided_difference_f1[0,:],evaluation_point_2)
start_time=time.time()
numerical_solution_f2=newton_polynomial(xi,divided_difference_f2[0,:],evaluation_point_2)
end_time=time.time()

error_f1_eva_point_1=np.abs(numerical_solution_f1 - analytical_solution_f_1_eva_point_2)
error_f2_eva_point_1=np.abs(numerical_solution_f2 - analytical_solution_f_2_eva_point_2)

print(f"N_points: {number_points_1}")
print(f"Evaluation Point : {evaluation_point_2}")
print(f"f_1(x) : cos(x)")
print(f"Numerical Solution : {numerical_solution_f1} , Analytical Solution : {analytical_solution_f_1_eva_point_2} , Error : {error_f1_eva_point_1}")

print(f"f_2(x) : 1/1+25x**2")
print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f_2_eva_point_2} , Error : {error_f2_eva_point_1}")

computed_time=end_time - start_time
print(f"Elasped time for Newton Polynomial at x_eval : {evaluation_point_1} = {computed_time} seconds")


#for n=25 and evaluation point 0

fi_1=f_1(qi)
fi_2=f_2(qi)

# Now,forming the divided diff table:

divided_difference_f1 =np.zeros((np.size(qi),np.size(qi)))
divided_difference_f2 =np.zeros((np.size(qi),np.size(qi)))
divided_difference_f1[:, 0]=fi_1
divided_difference_f2[:, 0]=fi_2


#from 1st dd onwards:

for i in range(1,np.size(qi)):
  prev_dd= i-1
  divided_difference_f1[:len(qi) - i, i]=  divided_difference_comp(qi, divided_difference_f1[:,prev_dd], i)
  divided_difference_f2[:len(qi) - i, i] = divided_difference_comp(qi, divided_difference_f2[:, prev_dd], i)

#for n=25 and evaluation point 0

numerical_solution_f1=newton_polynomial(qi,divided_difference_f1[0,:],evaluation_point_1)
start_time=time.time()
numerical_solution_f2=newton_polynomial(qi,divided_difference_f2[0,:],evaluation_point_1)
end_time=time.time()

error_f1_eva_point_1=np.abs(numerical_solution_f1 - analytical_solution_f_1_eva_point_1)
error_f2_eva_point_1=np.abs(numerical_solution_f2 - analytical_solution_f_2_eva_point_1)

print(f"N_points: {number_points_2}")
print(f"Evaluation Point : {evaluation_point_1}")
print(f"f_1(x) : cos(x)")
print(f"Numerical Solution : {numerical_solution_f1} , Analytical Solution : {analytical_solution_f_1_eva_point_1} , Error : {error_f1_eva_point_1}")

print(f"f_2(x) : 1/1+25x**2")
print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f_2_eva_point_1} , Error : {error_f2_eva_point_1}")

computed_time=end_time - start_time
print(f"Elasped time for Newton Polynomial at x_eval : {evaluation_point_1} = {computed_time} seconds")

#Now,for n=25 and evaluation point 0.95

numerical_solution_f1=newton_polynomial(qi,divided_difference_f1[0,:],evaluation_point_2)
start_time=time.time()
numerical_solution_f2=newton_polynomial(qi,divided_difference_f2[0,:],evaluation_point_2)
end_time=time.time()

error_f1_eva_point_1=np.abs(numerical_solution_f1 - analytical_solution_f_1_eva_point_2)
error_f2_eva_point_1=np.abs(numerical_solution_f2 - analytical_solution_f_2_eva_point_2)

print(f"N_points: {number_points_2}")
print(f"Evaluation Point : {evaluation_point_2}")
print(f"f_1(x) : cos(x)")
print(f"Numerical Solution : {numerical_solution_f1} , Analytical Solution : {analytical_solution_f_1_eva_point_2} , Error : {error_f1_eva_point_1}")

print(f"f_2(x) : 1/1+25x**2")
print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f_2_eva_point_2} , Error : {error_f2_eva_point_1}")

computed_time=end_time - start_time
print(f"Elasped time for Newton Polynomial at x_eval : {evaluation_point_1} = {computed_time} seconds")

#for n =50 and evaluation point 0

fi_1=f_1(ri)
fi_2=f_2(ri)

# Now,forming the divided diff table:

divided_difference_f1 =np.zeros((np.size(ri),np.size(ri)))
divided_difference_f2 =np.zeros((np.size(ri),np.size(ri)))
divided_difference_f1[:, 0]=fi_1
divided_difference_f2[:, 0]=fi_2


#from 1st dd onwards:

for i in range(1,np.size(ri)):
  prev_dd= i-1
  divided_difference_f1[:len(ri) - i, i]=  divided_difference_comp(ri, divided_difference_f1[:,prev_dd], i)
  divided_difference_f2[:len(ri) - i, i] = divided_difference_comp(ri, divided_difference_f2[:, prev_dd], i)

#for n=25 and evaluation point 0

numerical_solution_f1=newton_polynomial(ri,divided_difference_f1[0,:],evaluation_point_1)
start_time=time.time()
numerical_solution_f2=newton_polynomial(ri,divided_difference_f2[0,:],evaluation_point_1)
end_time=time.time()

error_f1_eva_point_1=np.abs(numerical_solution_f1 - analytical_solution_f_1_eva_point_1)
error_f2_eva_point_1=np.abs(numerical_solution_f2 - analytical_solution_f_2_eva_point_1)

print(f"N_points: {number_points_3}")
print(f"Evaluation Point : {evaluation_point_1}")
print(f"f_1(x) : cos(x)")
print(f"Numerical Solution : {numerical_solution_f1} , Analytical Solution : {analytical_solution_f_1_eva_point_1} , Error : {error_f1_eva_point_1}")

print(f"f_2(x) : 1/1+25x**2")
print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f_2_eva_point_1} , Error : {error_f2_eva_point_1}")

computed_time=end_time - start_time
print(f"Elasped time for Newton Polynomial at x_eval : {evaluation_point_1} = {computed_time} seconds")

#Now,for n=50 and evaluation point 0.95

numerical_solution_f1=newton_polynomial(ri,divided_difference_f1[0,:],evaluation_point_2)
start_time=time.time()
numerical_solution_f2=newton_polynomial(ri,divided_difference_f2[0,:],evaluation_point_2)
end_time=time.time()

error_f1_eva_point_1=np.abs(numerical_solution_f1 - analytical_solution_f_1_eva_point_2)
error_f2_eva_point_1=np.abs(numerical_solution_f2 - analytical_solution_f_2_eva_point_2)

print(f"N_points: {number_points_3}")
print(f"Evaluation Point : {evaluation_point_2}")
print(f"f_1(x) : cos(x)")
print(f"Numerical Solution : {numerical_solution_f1} , Analytical Solution : {analytical_solution_f_1_eva_point_2} , Error : {error_f1_eva_point_1}")

print(f"f_2(x) : 1/1+25x**2")
print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f_2_eva_point_2} , Error : {error_f2_eva_point_1}")

computed_time=end_time - start_time
print(f"Elasped time for Newton Polynomial at x_eval : {evaluation_point_1} = {computed_time} seconds")











```
