```python
import numpy as np
import time
def divided_difference_comp(xi, prev_divided_difference,i):
  size_div_difference=np.size(prev_divided_difference) -i       #most crucial step takes care of indexing
  divided_difference=np.zeros(size_div_difference)
  for j in range(0,size_div_difference):
    numerator=prev_divided_difference[j+1]-prev_divided_difference[j]
    denominator=xi[j+i]-xi[j]
    divided_difference[j]= numerator/denominator
  return divided_difference

def newton_polynomial(xi,divided_difference,evaluation_point):
  polynomial=divided_difference[0]                             #here divided diff contains the north-east edge point of the divided diff table
  basis_term=np.ones(np.size(xi))
  for i in range(1,np.size(xi)):
    for j in range(0,i):
      basis_term[i] *=evaluation_point - xi[j]
    polynomial += divided_difference[i]* basis_term[i]
  return polynomial

def first_hermite_divided_diff_computation(zi, prev_divided_difference, der_fi):
    size_divided_difference = len(prev_divided_difference) - 1
    divided_difference = np.zeros(size_divided_difference)
    for j in range(size_divided_difference):
        if zi[j] == zi[j+1]:
            index = j//2
            divided_difference[j] = der_fi[index]
        else:
            numerator = prev_divided_difference[j+1] - prev_divided_difference[j]
            denominator = zi[j+1] - zi[j]
            divided_difference[j] = numerator/denominator

    return divided_difference

#a=[8.3,8.3,8.6,8.6]
#b=[17.56492,17.56492,18.5075,18.5075]
#e=[3.116256,3.1341,3.151762]
#print(divided_difference_comp(b,e,c))


def f_1(x):
  return np.cos(x)

def f_2(x):
  return 1/(1+ 25*(x*x))

def der_f1(x):
  return -np.sin(x)

def der_f2(x):
  return (-50 * x)/((1+(25*x*x))**2)

#def der_f3(x):
  #return np.log(2.71*x)
#print(first_hermite_divided_diff_computation(b,e,der_f3))

if __name__ == "__main__":
  start_time=time.time()
  a=-1
  b=1
  number_points1=5
  number_points2=10
  xi=np.linspace(a,b,number_points1)
  yi=np.linspace(a,b,number_points2)
  fi_1=f_1(xi)
  fi_2=f_2(xi)
  der_fi_1=der_f1(xi)
  der_fi_2=der_f2(xi)

  evaluation_point_1=0
  evaluation_point_2=0.95


#for evaluation_point 0 and n=5


  analytical_solution_f1= f_1(evaluation_point_1)
  analytical_solution_f2= f_2(evaluation_point_1)

  zi= np.zeros(2*number_points1)
  f_zi_1=np.zeros(2*number_points1)
  f_zi_2=np.zeros(2*number_points1)
  for i in range(np.size(xi)):
    zi[2*i]=xi[i]
    zi[2*i+1]=xi[i]
    f_zi_1[2*i]=fi_1[i]    #change
    f_zi_1[2*i+1]=fi_1[i]
    f_zi_2[2*i]=fi_2[i]
    f_zi_2[2*i+1]=fi_2[i]


#divided difference table
  divided_difference_f1 = np.zeros((np.size(zi), np.size(zi)))
  divided_difference_f2 = np.zeros((np.size(zi), np.size(zi)))

# Oth div diff

  divided_difference_f1[:, 0] = f_zi_1
  divided_difference_f2[:, 0] = f_zi_2

#first div diff
  divided_difference_f1[:len(zi) - 1, 1] = first_hermite_divided_diff_computation(zi, divided_difference_f1[:, 0], der_fi_1)
  divided_difference_f2[:len(zi) - 1, 1] = first_hermite_divided_diff_computation(zi, divided_difference_f2[:, 0], der_fi_2)

# 2nd DD onwards

  for i in range(2, np.size(zi)):
        prev_dd = i-1;
        divided_difference_f1[:len(zi) - i, i] = divided_difference_comp(zi, divided_difference_f1[:, prev_dd], i)
        divided_difference_f2[:len(zi) - i, i] = divided_difference_comp(zi, divided_difference_f2[:, prev_dd], i)

  numerical_solution_f1 = newton_polynomial(zi, divided_difference_f1[0], evaluation_point_1)
  start_time = time.time()

  #numerical_solution_f2= newton_polynomial(evaluation_point_1, zi, divided_difference_f2)
  numerical_solution_f2 = newton_polynomial(zi, divided_difference_f2[0], evaluation_point_1)
  end_time =   time.time()

  error_f1 = np.abs(analytical_solution_f1 - numerical_solution_f1)
  error_f2 = np.abs(analytical_solution_f2 - numerical_solution_f2)

  print(f"N_points         : {number_points1}")
  print(f"Evaluation Point : {evaluation_point_1}")
  print(f"f_1(x) : cos(x)")
  print(f"Numerical Solution : {numerical_solution_f1}, Analytical Solution : {analytical_solution_f1}, Error : {error_f1}")
  print(f"f_2(x) : 1/(1+25x**2)")
  print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f2}, Error : {error_f2}")
  print(f"Elapsed time for Newton Method at x_eval : {evaluation_point_1} = {end_time - start_time} seconds")


#for n=5 and evaluation point 0.95


  analytical_solution_f1= f_1(evaluation_point_2)
  analytical_solution_f2= f_2(evaluation_point_2)
  numerical_solution_f1 = newton_polynomial(zi, divided_difference_f1[0], evaluation_point_2)
  start_time = time.time()
  #numerical_solution_f2= newton_polynomial(evaluation_point_1, zi, divided_difference_f2)
  numerical_solution_f2 = newton_polynomial(zi, divided_difference_f2[0], evaluation_point_2)
  end_time =   time.time()

  error_f1 = np.abs(analytical_solution_f1 - numerical_solution_f1)
  error_f2 = np.abs(analytical_solution_f2 - numerical_solution_f2)

  print(f"N_points         : {number_points1}")
  print(f"Evaluation Point : {evaluation_point_2}")
  print(f"f_1(x) : cos(x)")
  print(f"Numerical Solution : {numerical_solution_f1}, Analytical Solution : {analytical_solution_f1}, Error : {error_f1}")
  print(f"f_2(x) : 1/(1+25x**2)")
  print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f2}, Error : {error_f2}")
  print(f"Elapsed time for Newton Method at x_eval : {evaluation_point_2} = {end_time - start_time} seconds")


  ##########################################################################


  fi_1=f_1(yi)
  fi_2=f_2(yi)
  der_fi_1=der_f1(yi)
  der_fi_2=der_f2(yi)

  evaluation_point_1=0
  evaluation_point_2=0.95


#for evaluation_point 0 and n=10


  analytical_solution_f1= f_1(evaluation_point_1)
  analytical_solution_f2= f_2(evaluation_point_1)

  zi= np.zeros(2*number_points2)
  f_zi_1=np.zeros(2*number_points2)
  f_zi_2=np.zeros(2*number_points2)
  for i in range(np.size(yi)):
    zi[2*i]=yi[i]
    zi[2*i+1]=yi[i]
    f_zi_1[2*i]=fi_1[i]
    f_zi_1[2*i+1]=fi_1[i]
    f_zi_2[2*i]=fi_2[i]
    f_zi_2[2*i+1]=fi_2[i]


#divided difference table
  divided_difference_f1 = np.zeros((np.size(zi), np.size(zi)))
  divided_difference_f2 = np.zeros((np.size(zi), np.size(zi)))

# Oth div diff

  divided_difference_f1[:, 0] = f_zi_1
  divided_difference_f2[:, 0] = f_zi_2

#first div diff
  divided_difference_f1[:len(zi) - 1, 1] = first_hermite_divided_diff_computation(zi, divided_difference_f1[:, 0], der_fi_1)
  divided_difference_f2[:len(zi) - 1, 1] = first_hermite_divided_diff_computation(zi, divided_difference_f2[:, 0], der_fi_2)

# 2nd DD onwards

  for i in range(2, np.size(zi)):
        prev_dd = i-1;
        divided_difference_f1[:len(zi) - i, i] = divided_difference_comp(zi, divided_difference_f1[:, prev_dd], i)
        divided_difference_f2[:len(zi) - i, i] = divided_difference_comp(zi, divided_difference_f2[:, prev_dd], i)

  numerical_solution_f1 = newton_polynomial(zi, divided_difference_f1[0], evaluation_point_1)
  start_time = time.time()
  #numerical_solution_f2= newton_polynomial(evaluation_point_1, zi, divided_difference_f2)
  numerical_solution_f2 = newton_polynomial(zi, divided_difference_f2[0], evaluation_point_1)
  end_time =   time.time()

  error_f1 = np.abs(analytical_solution_f1 - numerical_solution_f1)
  error_f2 = np.abs(analytical_solution_f2 - numerical_solution_f2)

  print(f"N_points         : {number_points2}")
  print(f"Evaluation Point : {evaluation_point_1}")
  print(f"f_1(x) : cos(x)")
  print(f"Numerical Solution : {numerical_solution_f1}, Analytical Solution : {analytical_solution_f1}, Error : {error_f1}")
  print(f"f_2(x) : 1/(1+25x**2)")
  print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f2}, Error : {error_f2}")
  print(f"Elapsed time for Newton Method at x_eval : {evaluation_point_1} = {end_time - start_time} seconds")


  #for evaluation point 0.95 and n=5
  analytical_solution_f1= f_1(evaluation_point_2)
  analytical_solution_f2= f_2(evaluation_point_2)

  numerical_solution_f1 = newton_polynomial(zi, divided_difference_f1[0], evaluation_point_2)
  start_time = time.time()
  #numerical_solution_f2= newton_polynomial(evaluation_point_1, zi, divided_difference_f2)
  numerical_solution_f2 = newton_polynomial(zi, divided_difference_f2[0], evaluation_point_2)
  end_time =   time.time()

  error_f1 = np.abs(analytical_solution_f1 - numerical_solution_f1)
  error_f2 = np.abs(analytical_solution_f2 - numerical_solution_f2)

  print(f"N_points         : {number_points2}")
  print(f"Evaluation Point : {evaluation_point_2}")
  print(f"f_1(x) : cos(x)")
  print(f"Numerical Solution : {numerical_solution_f1}, Analytical Solution : {analytical_solution_f1}, Error : {error_f1}")
  print(f"f_2(x) : 1/(1+25x**2)")
  print(f"Numerical Solution : {numerical_solution_f2}, Analytical Solution : {analytical_solution_f2}, Error : {error_f2}")
  print(f"Elapsed time for Newton Method at x_eval : {evaluation_point_2} = {end_time - start_time} seconds")









```
